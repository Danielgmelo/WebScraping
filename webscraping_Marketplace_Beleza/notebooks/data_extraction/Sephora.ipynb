```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
import json
import datetime
from time import sleep
import logging

logging.basicConfig(level=logging.INFO)
```

```python
driver = webdriver.Chrome("C:/Users/daniel.melo/projetos/StudyGroup/chromedriver_win32/chromedriver.exe")
driver.get("https://www.sephora.com.br/perfumes/feminino")
logging.info(f'Página {driver.current_url} acessada')
```

```python
count=0
while count <=10:
    button_more_products = (driver.find_element_by_xpath('//*[@id="see_more_products"]'))
    button_more_products.click()
    sleep(3)
    logging.info(f'Mais produtos, página {count}')
    count = count + 1
```

```python
content = driver.page_source
soup = bs(content, 'html.parser')
```

```python
day = datetime.datetime.now().strftime('%x')
hour = datetime.datetime.now().strftime('%X')
date = datetime.datetime.now()

items = soup.find_all('li', class_ = 'item')

sku = []
descricao = []
gross_price = []
price = []
pagto = [] 
    
for i in items:
    sku_ = i['data-productid']
    
    count_out_of_stok = 0
    if i['class'] == ['item', 'out-of-stock']:
        logging.info(f'{sku_}: out-of-stock')
        count_out_of_stok += 1
        continue
        
    logging.info(f'Dados do SKU {sku_}')
    
    sku.append(
        i['data-productid'])
    descricao.append(
        i.find('a', class_ = "product-image")['title'])
    gross_price.append(
        i['data-product-price'])
    price.append(
        i['data-product-final-price'])
    pagto.append(
        i.find('div', class_ = "installments").text)

logging.info(f'Qtde de out of stock: {count_out_of_stok}')

base = (pd.DataFrame({'sku':sku,
                      'descricao':descricao,
                      'gross_price':gross_price,
                      'price':price,
                      'pagamento': pagto,
                      'date':date,
                      'day':day,
                      'hour':hour})
        )

with open('../data/sephora.csv', 'w') as f:
    base.to_csv(f, header=f.tell()==0, index=False, decimal='.', encoding='ISO-8859-1')

driver.close()
#Há algns casos em que o frete tem *
```
